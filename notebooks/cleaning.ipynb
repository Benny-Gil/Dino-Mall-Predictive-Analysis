# Activity 1 Cleaning Process

## Loading the dataset

```python
import numpy as np
import pandas as pd

dataframe = pd.read_csv('../data/dino_mall_raw.csv')
dataframe
```

## Dropping Rows based on consent
Dropping rows where consent is NA since this would mean either that we failed to acquire their consent, or that they did not consent to the use of their data

```python
print(f"Number of no consent: {dataframe['consent'].isnull().sum()}")


dataframe = dataframe.dropna(subset='consent')
dataframe
```

## Removing invalid age ranges
Some rows are dates but are missing year, additionally this they are all `19-Oct`, I will be assuming this because of incorrect parsing. Therefore i will be removing these rows since we cannot use this data when associating age brackets, this is inconsistent data and will be noisy

```python
# This will keep only rows where values follow the pattern like "20-29", "30-39", etc.
dataframe = dataframe[dataframe['age_range'].str.match(r'^\d+-\d+$')]
dataframe
```

## Substituting values of monthly income
There are some blank values, since we have removed the rows with no consent, we will be assuming that NA values will be `Prefer to not say`

```python
dataframe['monthly_income'] = dataframe['monthly_income'].fillna("Prefer to not say")
dataframe
```

## Printing the output

```python
dataframe.to_csv('../data/dino_mall_cleaned.csv', index=False)
```
